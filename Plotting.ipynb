```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,1000)
y = 1/(1 + x**2/2 * (1+ 2/x**2 - np.sqrt(1 + 4/x**2)))

fig, ax = plt.subplots()

ax.set_title(r"$A_a$ as a function of the parameters")
ax.set_xlabel(r"$y = \frac{V_1 - V_2^{a} - t_{c}^{a}}{t_{c}^{b}}$")#
ax.set_ylabel(r"$A_a(y)$")
ax.grid()
#ax.set_ylim((-5, 5))
#  ax.legend()
ax.plot(x, y)


plt.tight_layout()

plt.savefig("A_a-of-y"+".svg", dpi=300)
plt.show()
```
